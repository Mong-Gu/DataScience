### 레버리지가 높은 이상치 제거 과정

```python
influence = fit.get_influence() #영향도 정보객체를 구할 수 있다. 
hat = influence.hat_matrix_diag
```

```python
plt.figure(figsize=(10,2)) #가로, 세로의 길이설정
plt.stem(hat)
plt.axhline(0.02, c="g", ls="--")
plt.title("leverage for each data point")
plt.show()
```

```python
hat.sum() #레버리지 평균이 2/100=0.02
```

```python
cooks_d2, pvals = influence.cooks_distance
K=influence.k_vars
```

```python
fox_cr = 4/(len(data['attend'])-K-1)
```

```python
idx = np.where(cooks_d2 > fox_cr)[0]
idx
#cap 은 41, 54에서 yes 나머지는 왜 이상치인지 모르겠음
```

```python
new_data = data.drop(idx,0)
```

```python
new_fit=ols('attend ~ C(month) + C(day_of_week) + C(park)', data=new_data).fit()
new_fit.summary()
```
